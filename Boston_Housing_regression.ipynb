```python
# Importing the dataset along with the description and features
from sklearn.datasets import load_boston
X,y = load_boston(return_X_y=True)
Features = load_boston(return_X_y=False)['feature_names']
print(load_boston(return_X_y=False)['DESCR'])
```

```python
#Making a Data Frame to slice and dice data using Pandas Features
import pandas as pd
Data = pd.DataFrame(X,columns = Features)
Data['Target'] = y
```

```python
# building a test and train set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(Data.iloc[:,0:13],Data['Target'], test_size = 0.2, random_state = 313)
```

```python
# building a model with cross validation
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

Lin_reg = LinearRegression()
Lin_reg.fit(X_train,y_train)
```

```python
# Training the model using Cross validation
cross_val = cross_val_score (estimator = Lin_reg, X= X_train, y= y_train, cv= 5)
```

```python
# calculating R2 value for trainset and testset and comparing with the crossvalidation R2
Train_pred = Lin_reg.predict(X_train)
r2_train = r2_score(Train_pred, y_train)

Test_pred = Lin_reg.predict(X_test)
r2_test = r2_score(Test_pred,y_test)
```

```python
## printing the R2 value and understanding the performance
print ("""
        The R2 value of the crossvalidation is : {} 
        The R2 value of the Train set is  : {} 
        The R2 value of the Test set is : {} """.format(cross_val.mean(),r2_train,r2_test))
```

```python
## calculating the RMSE that is root mean squared error
import numpy as np
from sklearn.metrics import mean_squared_error
Error_Train = np.sqrt(mean_squared_error(y_train,Train_pred))
Error_Test = np.sqrt(mean_squared_error(y_test,Test_pred))
```

```python
print("The RMSE for Training Data is {} \nRMSE for test is {}".format(Error_Train,Error_Test))
```

```python
# if we engineer some features and use Regularisation. since our R2 value says that 
# there remains some relationship between the predictor variables and dependent Variables
# that is unexplainable.
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2 
poly_features = poly.fit_transform(X_train)
Poly_reg = LinearRegression()
Poly_reg.fit(poly_features,y_train)
poly_pred = Poly_reg.predict(poly.fit_transform(X_test))
r2_poly = r2_score(poly_pred,y_test)

```

```python
##using Regularisation
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
from sklearn.pipeline import Pipeline

Steps = [
        ('Scaler', StandardScaler()),('Poly', PolynomialFeatures()),('Reg_Lasso',Lasso(alpha = 0.002,
                fit_intercept=True, max_iter=250))]
Pipeline_Boston = Pipeline(Steps)
Pipeline_Boston.fit(X_train,y_train)
Lasso_pred = Pipeline_Boston.predict(X_test)
r2_Lasso = r2_score(Lasso_pred,y_test)
Error_Lasso = np.sqrt(mean_squared_error(y_test,Lasso_pred))
print("the r2 value is : {} and the RMSE is : {}".format(r2_Lasso,Error_Lasso))


```
