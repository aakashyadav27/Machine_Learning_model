```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression   
```

```python
data=pd.read_csv("startups.csv")
data.head()
```

```python
data.groupby("State").sum("Profit")
```

```python
dummies=pd.get_dummies(data.State)
dummies
```

```python
merged=pd.concat([data,dummies],axis='columns')
merged
```

```python
final=merged.copy()
final.head()
```

```python
final.drop(axis=1,columns=["State","New York"],inplace=True)
```

```python
final.head()
```

```python
x=final[["R&D Spend","Administration","Marketing Spend","California","Florida"]].values
y=final[["Profit"]].values
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=0)
test_y
```

```python
MLR=LinearRegression()
MLR.fit(train_x,train_y)
```

```python
pred_y=MLR.predict(test_x)
print([pred_y])
print([test_y])
```

```python
MLR.predict([[180000,151377.59,443898.53,1,0]])
```

```python
MLR.intercept_
```

```python
MLR.coef_
```

```python
a1=MLR.score(train_x,train_y)
b1=MLR.score(test_x,test_y)
a1-b1
```

# backward elimination

```python
import statsmodels.api as sm
```

```python
x=np.append(arr=np.ones((50,1)).astype(int),values=x,axis=1)
```

```python
x
```

```python
x_opt=x[:,[0,1,2,3,4,5]]
```

```python
reg_OLS=sm.OLS(endog=y,exog=x_opt).fit()
```

```python
reg_OLS.summary()
```

```python
x_opt=x[:,[0,1,3,4,5]]
reg_OLS=sm.OLS(endog=y,exog=x_opt).fit()
reg_OLS.summary()
```

```python
x_opt=x[:,[0,1,3,5]]
reg_OLS=sm.OLS(endog=y,exog=x_opt).fit()
reg_OLS.summary()
```

```python
x_opt=x[:,[0,1,3]]
reg_OLS=sm.OLS(endog=y,exog=x_opt).fit()
reg_OLS.summary()
```

```python
x_opt=x[:,[0,1]]
reg_OLS=sm.OLS(endog=y,exog=x_opt).fit()
reg_OLS.summary()
```

As we can see in the above output image, only two variables are left. So only the R&D independent variable is a significant variable for the prediction. So we can now predict efficiently using this variable

```python
data_set= pd.read_csv('startups.csv')  
  
data_set=data_set.iloc[:,[0,4]]
data_set.head()
```

```python
reg3=LinearRegression()
x=data_set.iloc[:,0].values.reshape(-1,1)
y=data_set.iloc[:,1].values.reshape(-1,1)
training_x,testing_x,training_y,testing_y=train_test_split(x,y,test_size=0.3,random_state=0)
reg3.fit(training_x,training_y)
```

```python
reg3.predict(testing_x)
```

```python
print(testing_y)
```

```python
a=reg3.score(training_x,training_y)
b=reg3.score(testing_x,testing_y)
print("before backward elimination=",a1-b1)
print("after backward elimination",a-b)
```
